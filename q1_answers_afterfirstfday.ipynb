```python
# Getting all auplicated
dupli_items = items[items.duplicated(subset='item_name', keep=False)]
dupli_items = dupli_items.copy() # For removing Error 

# Counting total values not null per row
dupli_items['total_not_null'] = dupli_items.notna().sum(axis=1)

# Sorting
dupli_items.sort_values(by=['item_name', 'total_not_null'], ascending=[True, False], inplace=True)

display(dupli_items)
```

```python
keep_items = dupli_items.drop_duplicates(subset='item_name', keep='first')

del_duplicates = dupli_items[~dupli_items['id'].isin(keep_items['id'])]

items_after_remove = items[~items['id'].isin(del_duplicates['id'])]

dupli_count_after_remove = items_after_remove.groupby('item_name')['id'].count()
dupli_count_after_remove[dupli_count_after_remove > 1]
```

```python
merged = del_duplicates.merge(keep_items, how='inner', suffixes=('_del', '_keep'), on='item_name')

result = dict(zip(merged['id_del'], merged['id_keep']))

print("Before removing")
display(items['id'].isin(del_duplicates['id']).sum())

items['id'] = items['id'].replace(result)

print("-----" *20)

print("After removing")
display(items['id'].isin(del_duplicates['id']).sum())
```

```python
customers_copy = customers.copy()


kept = customers_copy.dropna(axis=0, thresh=(customers_copy.shape[1] - 2))
removed = customers_copy[~customers_copy['id'].isin(kept['id'])]

customers.drop(removed.index, inplace=True)

removed_customer_id = list(removed['id'])
display(removed_customer_id)

removed_from_customers = orders[orders['customer_id'].isin(removed_customer_id)]

removed_customer_order_id = list(removed_from_customers['order_id'])
display(removed_customer_order_id)

orders_item_after_remove = orders_item[~orders_item['order_id'].isin(removed_customer_order_id)]
orders_item = orders_item_after_remove
```
